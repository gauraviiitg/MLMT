```python
import pandas as pd
import numpy as np
import glob
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# List of CSV files
csv_files = glob.glob('ML_Test_Data_*.csv')  # Update the file pattern

# Initialize arrays to store data for linear regression
fDelT_data = []
std_dev_fDelT = []
dist_data = []

# Load data and populate arrays
for file in csv_files:
    dist = int(file.split('_')[-1].split('.')[0])  # Extract the 'dist' value from the file name
    data = pd.read_csv(file)
    
    fDelT_data.append(data['fDelT'].mean())  # Use mean instead of mode as mentioned
    std_dev_fDelT.append(data['fDelT'].std())
    dist_data.append(dist)

# Convert arrays to numpy arrays
fDelT_data = np.array(fDelT_data).reshape(-1, 1)
std_dev_fDelT = np.array(std_dev_fDelT).reshape(-1, 1)
dist_data = np.array(dist_data)

# Create and fit the linear regression model
X_regression = np.hstack((fDelT_data, std_dev_fDelT))
model_regression = LinearRegression()
model_regression.fit(X_regression, dist_data)

# Get the coefficients of the linear regression model
coefficients = model_regression.coef_
intercept = model_regression.intercept_

# Calculate predictions using the linear regression model
dist_predicted = model_regression.predict(X_regression)

# Calculate Mean Squared Error
mse = mean_squared_error(dist_data, dist_predicted)
print("Mean Squared Error:", mse)

# Print the equation of the linear regression model
equation = f'Equation: dist = {intercept:.2f} + {coefficients[0]:.2f} * fDelT + {coefficients[1]:.2f} * std_dev_fDelT'
print(equation)

# Plot the data points with the regression line
plt.figure(figsize=(10, 6))
plt.scatter(fDelT_data, dist_data, color='blue', label='Data Points')
plt.plot(fDelT_data, dist_predicted, color='red', label='Regression Line')
plt.xlabel('fDelT')
plt.ylabel('dist')
plt.title('Linear Regression: Predicting dist from fDelT and std_dev_fDelT')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob

# Define the regression equation parameters
intercept = -1.63
coef_fDelT = -10.88
coef_std_dev = 1.21

# List all CSV files matching the pattern
file_list = glob.glob('ML_Test_Data_*.csv')

for file in file_list:
    # Load the CSV file into a pandas DataFrame
    df = pd.read_csv(file)
    
    # Calculate the predicted dist values using the regression equation
    predicted_dist = intercept + coef_fDelT * df['fDelT'] + coef_std_dev * np.std(df['fDelT'])
    
    # Calculate the standard deviation of the predicted dist values
    std_dev_predicted = np.std(predicted_dist)
    
    # Create a figure
    plt.figure(figsize=(8, 6))
    
    # Plot histogram
    plt.hist(predicted_dist, bins=20, alpha=0.7)
    
    # Add labels and title to the plot
    plt.xlabel('Predicted dist')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Predicted dist Values - {file}')
    
    # Add standard deviation of predicted values as a text annotation
    plt.text(0.95, 0.85, f'Std Dev: {std_dev_predicted:.2f}', transform=plt.gca().transAxes,
             verticalalignment='top', horizontalalignment='right', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))
    
    # Display the plot
    plt.show()
    
    # Save the plot as an image
    plt.savefig(f'histogram_{file}.png')
    
    # Clear the current plot for the next file
    plt.clf()

# Show a message when all graphs are saved and shown
print("Histograms saved and displayed successfully.")
```
