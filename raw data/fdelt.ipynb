```python
import pandas as pd
import matplotlib.pyplot as plt

# List of file names to read
file_names = ['Data_0.csv', 'Data_+10.csv', 'Data_+20.csv', 'Data_+30.csv', 'Data_+40.csv', 'Data_+45.csv',
              'Data_-10.csv', 'Data_-20.csv', 'Data_-30.csv', 'Data_-40.csv', 'Data_-45.csv']

# List to store the averages for each file
average_values = []

# Loop through each file, read the data, and calculate the average of fDelt
for file_name in file_names:
    # Load the data from the CSV file
    data = pd.read_csv(file_name)
    
    # Calculate the average of fDelt for the current file
    average_fDelt = data['am'].mean()
    
    # Append the average to the list
    average_values.append(average_fDelt)

# Plot the average values for each file
plt.figure(figsize=(10, 6))
plt.bar(file_names, average_values)
plt.xlabel('File Names')
plt.ylabel('Average fDelt')
plt.title('Average fDelt for each file')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Add values above each bar
for i, v in enumerate(average_values):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom')

# Save the plot as a PNG file
plt.savefig('average_fDelt_per_file.png', dpi=300, bbox_inches='tight')

# Show the plot (optional)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of file names to read
file_names = ['Data_0.csv', 'Data_+10.csv', 'Data_+20.csv', 'Data_+30.csv', 'Data_+40.csv', 'Data_+45.csv',
              'Data_-10.csv', 'Data_-20.csv', 'Data_-30.csv', 'Data_-40.csv', 'Data_-45.csv']

# List to store the averages for each file
average_values = []

# Loop through each file, read the data, remove outliers, and calculate the average of fDelt
for file_name in file_names:
    # Load the data from the CSV file
    data = pd.read_csv(file_name)
    
    # Calculate the 25th and 75th percentiles (Q1 and Q3)
    Q1 = data['am'].quantile(0.25)
    Q3 = data['am'].quantile(0.75)
    
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    
    # Determine the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Remove outliers from the data
    data = data[(data['am'] >= lower_bound) & (data['am'] <= upper_bound)]
    
    # Calculate the average of fDelt for the current file
    average_fDelt = data['am'].mean()
    
    # Append the average to the list
    average_values.append(average_fDelt)

# Plot the average values for each file
plt.figure(figsize=(10, 6))
plt.bar(file_names, average_values)
plt.xlabel('File Names')
plt.ylabel('Average fDelt')
plt.title('Average fDelt for each file')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Add values above each bar
for i, v in enumerate(average_values):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom')

# Save the plot as a PNG file
plt.savefig('average_fDelt_per_file.png', dpi=300, bbox_inches='tight')

# Show the plot (optional)
plt.show()
```
