```python
#linear regression
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# List of files
files = [
    "output0.csv",
    "output+10.csv",
    "output+20.csv",
    "output+30.csv",
    "output+40.csv",
    "output+45.csv",
    "output-10.csv",
    "output-20.csv",
    "output-30.csv",
    "output-40.csv",
    "output-45.csv"
]

# Initialize lists to store training and testing data
train_data = []
test_data = []

# Load data from each file
for file in files:
    df = pd.read_csv(file)
    
    # Randomly select 10% of data as testing data
    train, test = train_test_split(df, test_size=0.1, random_state=42)
    
    # Append data to the respective lists
    train_data.append(train)
    test_data.append(test)

# Concatenate all data from different files
train_data = pd.concat(train_data)
test_data = pd.concat(test_data)

# Define features and target variables
features = ['c']
target = 'dist'

# Create and fit the linear regression model
model = LinearRegression()
model.fit(train_data[features], train_data[target])

# Predict on the testing data
predictions = model.predict(test_data[features])

# Calculate the absolute mean difference
abs_mean_diff = abs(predictions - test_data[target]).mean()

# Calculate the R-squared value
r_squared = r2_score(test_data[target], predictions)

# Calculate the Mean Squared Error
mse = mean_squared_error(test_data[target], predictions)

# Calculate the difference between actual and predicted values
difference = test_data[target] - predictions

# Plot a histogram of the difference
plt.hist(difference, bins='auto', alpha=0.7, color='green')
plt.xlabel('Difference')
plt.ylabel('Frequency')
plt.title('Histogram of Difference between Actual and Predicted')
plt.show()

print("Absolute Mean Difference:", abs_mean_diff)
print("R-squared:", r_squared)
print("MSE:", mse)
```

```python
#poly regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# List of input files
input_files = [
    "output0.csv",
    "output+10.csv",
    "output+20.csv",
    "output+30.csv",
    "output+40.csv",
    "output+45.csv",
    "output-10.csv",
    "output-20.csv",
    "output-30.csv",
    "output-40.csv",
    "output-45.csv"
]

best_degree = None
best_abs_mean_diff = float('inf')
best_r_squared = -float('inf')
best_mse = float('inf')

# Loop over different polynomial degrees
for degree in range(1, 12):
    abs_mean_diff_list = []
    r_squared_list = []
    mse_list = []
    difference_list = []

    # Load the data from each input file
    data = []
    for file in input_files:
        df = pd.read_csv(file)
        data.append(df)

    # Combine all data into a single DataFrame
    combined_data = pd.concat(data, ignore_index=True)

    # Extract the input features (c) and target variable (dist)
    X = combined_data[["c"]]
    y = combined_data["dist"]

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

    # Perform polynomial feature transformation
    poly_features = PolynomialFeatures(degree=degree)
    X_train_poly = poly_features.fit_transform(X_train)
    X_test_poly = poly_features.transform(X_test)

    # Perform polynomial regression
    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Predict the target variable for test set
    y_test_pred = model.predict(X_test_poly)

    # Calculate the Absolute Mean Difference
    abs_mean_diff = mean_absolute_error(y_test, y_test_pred)

    # Calculate the R-squared value
    r_squared = r2_score(y_test, y_test_pred)

    # Calculate the Mean Squared Error
    mse = mean_squared_error(y_test, y_test_pred)

    # Calculate the difference between actual and predicted values
    difference = y_test - y_test_pred

    abs_mean_diff_list.append(abs_mean_diff)
    r_squared_list.append(r_squared)
    mse_list.append(mse)
    difference_list.append(difference)

    # Determine if the current degree gives better results
    if abs_mean_diff < best_abs_mean_diff:
        best_degree = degree
        best_abs_mean_diff = abs_mean_diff
        best_r_squared = r_squared
        best_mse = mse

    print(f"Degree: {degree}")
    print("Absolute Mean Difference:", abs_mean_diff)
    print("R-squared:", r_squared)
    print("MSE:", mse)
    print("")

    # Plot histogram of the difference between actual and predicted values
    plt.hist(difference, bins='auto', alpha=0.7, color='green')
    plt.xlabel('Difference')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Difference (Degree {degree})')
    plt.show()

print("Best Results:")
print(f"Degree: {best_degree}")
print("Best Absolute Mean Difference:", best_abs_mean_diff)
print("Best R-squared:", best_r_squared)
print("Best MSE:", best_mse)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# List of input files
input_files = [
    "output0.csv",
    "output+10.csv",
    "output+20.csv",
    "output+30.csv",
    "output+40.csv",
    "output+45.csv",
    "output-10.csv",
    "output-20.csv",
    "output-30.csv",
    "output-40.csv",
    "output-45.csv"
]

# Load the data from each input file
data = []
for file in input_files:
    df = pd.read_csv(file)
    data.append(df)

# Combine all data into a single DataFrame
combined_data = pd.concat(data, ignore_index=True)

# Extract the input features (c) and target variable (dist)
X = combined_data[["c"]]
y = combined_data["dist"]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Perform multiple regression
model = LinearRegression()
model.fit(X_train, y_train)

# Predict the target variable for training and test sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate the Absolute Mean Difference
absolute_mean_difference = mean_absolute_error(y_test, y_test_pred)

# Calculate the Mean Squared Error
mse = mean_squared_error(y_test, y_test_pred)

# Calculate the R-squared value
r_squared = r2_score(y_test, y_test_pred)

# Calculate the difference between actual and predicted values
difference = y_test - y_test_pred

# Plot a histogram of the difference

plt.hist(difference, bins='auto', alpha=0.7, color='green')
plt.xlabel('Difference')
plt.ylabel('Frequency')
plt.title('Histogram of Difference between Actual and Predicted')
plt.show()

# Print the results
print("Absolute Mean Difference:", absolute_mean_difference)
print("Mean Squared Error:", mse)
print("R-squared:", r_squared)

```

```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# List of file names
file_names = [
    "Data_-45",
    "Data_-40",
    "Data_-30",
    "Data_-20",
    "Data_-10",
    "Data_0",
    "Data_+10",
    "Data_+20",
    "Data_+30",
    "Data_+40",
    "Data_+45",
]

# Initialize lists to store mean fTstamp values and file name indices
mean_fTstamp = []
file_indices = []

# Iterate over the file names and assign indices
for i, file_name in enumerate(file_names):
    # Open the ROOT file and access the tree
    file = uproot.open(file_name + ".root")
    tree = file["ftree"]

    # Read the "fTstamp" branch as an array
    fTstamp_values = tree["fDelt"].array()

    # Calculate the mean of fTstamp using NumPy
    mean = np.mean(fTstamp_values)
    mean_fTstamp.append(mean)
    file_indices.append(i)

# Convert the lists to NumPy arrays
file_indices = np.array(file_indices)
mean_fTstamp = np.array(mean_fTstamp)

# Perform polynomial regression with degrees up to 21
degrees = range(1, 22)
r2_scores = []

for degree in degrees:
    # Perform polynomial feature transformation
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(file_indices[:, np.newaxis])

    # Perform linear regression
    model = LinearRegression()
    model.fit(X_poly, mean_fTstamp)

    # Predict the mean fTstamp using the model
    mean_fTstamp_pred = model.predict(X_poly)

    # Calculate the R-squared score
    r2 = r2_score(mean_fTstamp, mean_fTstamp_pred)
    r2_scores.append(r2)

# Find the degree with the highest R-squared score
best_degree = degrees[np.argmax(r2_scores)]

# Perform polynomial regression with the best degree
poly_features = PolynomialFeatures(degree=best_degree)
X_poly = poly_features.fit_transform(file_indices[:, np.newaxis])
model = LinearRegression()
model.fit(X_poly, mean_fTstamp)
mean_fTstamp_pred = model.predict(X_poly)

# Plot the graph
plt.plot(file_names, mean_fTstamp, marker='o', label='Actual')
plt.plot(file_names, mean_fTstamp_pred, marker='o', label='Predicted')
plt.xlabel("File Name")
plt.ylabel("Mean fTstamp")
plt.title("Mean fTstamp for Different File Names")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Print the equation
equation = f"Mean fTstamp = "
coefficients = model.coef_
for i, coeff in enumerate(coefficients):
    if i == 0:
        equation += f"{coeff:.2f}"
    else:
        equation += f" + {coeff:.2f} * x^{i}"

print("Equation:", equation)
```
