```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

train_df = pd.read_csv('itraining.csv')
test_df = pd.read_csv('itesting.csv')

# Linear features
linear_features =[  'fDelT','corrected',]  #'fDelT',  'fQMean', 'corrected', 'fqratio'

# Polynomial features
poly_features = [ 'fQMean', 'fqratio'] 

train_X_linear = train_df[linear_features]
train_X_poly = train_df[poly_features]
train_y = train_df['dist']

test_X_linear = test_df[linear_features]
test_X_poly = test_df[poly_features]
test_y = test_df['dist']

# Create polynomial features for the polynomial variables
degree = 2
poly_transformer = PolynomialFeatures(degree=degree)
train_X_poly = poly_transformer.fit_transform(train_X_poly)
test_X_poly = poly_transformer.transform(test_X_poly)

# Combine linear and polynomial features
train_X = np.concatenate((train_X_linear, train_X_poly), axis=1)
test_X = np.concatenate((test_X_linear, test_X_poly), axis=1)

# Create and fit the mixed degree polynomial regression model
model = LinearRegression()
model.fit(train_X, train_y)

# Make predictions using the polynomial regression model
predictions = model.predict(test_X)

difference = predictions - test_y
average_difference = difference.abs().mean()
print("Average Difference:", average_difference)

mse = mean_squared_error(test_y, predictions)
mae = mean_absolute_error(test_y, predictions)
r2 = r2_score(test_y, predictions)

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared:", r2)

plt.figure(figsize=(10, 6))
plt.hist(difference, bins=20, edgecolor='black')
plt.xlabel('Difference (Predicted - Actual)')
plt.ylabel('Frequency')
plt.title('Distribution of Difference between Predicted and Actual Values')
plt.grid(True)
plt.show()
```
