```python
import pandas as pd
import matplotlib.pyplot as plt

# List of CSV filenames
csv_files = [
    "ML_Data_0.csv", "ML_Data_10.csv", "ML_Data_20.csv", "ML_Data_30.csv", "ML_Data_40.csv",
    "ML_Data_-10.csv", "ML_Data_-20.csv", "ML_Data_-30.csv", "ML_Data_-40.csv"
]

# Load and process each CSV file
for csv_file in csv_files:
    # Load the CSV file
    df = pd.read_csv(csv_file)
    
    # Calculate the average of the "fDelT" column
    avg_fDelT = df['fDelT'].mean()
    
    # Create a histogram
    plt.figure(figsize=(8, 6))
    plt.hist(df['fDelT'], bins=100, color='blue', alpha=0.7)
    plt.axvline(avg_fDelT, color='red', linestyle='dashed', linewidth=2, label=f'Avg: {avg_fDelT:.2f}')
    plt.xlabel('fDelT')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of fDelT - {csv_file}')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    print(f"Average fDelT for {csv_file}: {avg_fDelT:.2f}")
```
