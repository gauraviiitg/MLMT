```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

csv_files = ['Data_+45.csv', 'Data_-45.csv']

for file in csv_files:
    # 1. Reading the CSV file
    df = pd.read_csv(file)
    
    # 2. Calculating std_dev_fDelT
    std_dev_fDelT = df['corrected'].std()

    # 3. Applying the equation
    dist = (-55.79589 + 0.00165 * df['corrected'] + 
           0.03792 * std_dev_fDelT - 
           0.00001 * (std_dev_fDelT)**2)
    
    # 4. Generating histogram and calculating its standard deviation
    plt.hist(dist, bins=150) # Adjust bins as necessary
    plt.title(f'Histogram for {file}')
    plt.xlabel('Position(in cm)')
    plt.ylabel('Frequency')
    plt.show()

    hist_std_dev = np.std(dist, ddof=1)
    print(f"Standard deviation of histogram for {file}: {hist_std_dev}")
```
